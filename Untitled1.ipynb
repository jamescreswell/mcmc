```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_x = np.linspace(0, 10, num=1000)

theta_true = [-2.5, 4.5]

def model(x, theta):
    return theta[0] * x + theta[1]

data_y = model(data_x, theta_true)

data_x = np.random.normal(data_x, 1)
data_y = np.random.normal(data_y, 1)
```

```python
plt.plot(data_x, data_y, 'k.')
```

```python
def log_prior(theta):
    m, b, sigma = theta
    if sigma < 0:
        return -np.inf
    return -1.5 * np.log(1 + m ** 2) - np.log(sigma)

def log_likelihood(data, theta):
    data_x, data_y = data
    m, b, sigma = theta
    model_y = model(data_x, [m, b])
    return -0.5*np.sum(np.log(2 * np.pi * sigma**2) + (data_y - model_y)**2 / sigma**2)
```

```python
from mcmc import Metropolis

x = Metropolis([data_x, data_y], log_prior, log_likelihood)

x.iterate([5.5, 4.5, 1.0], 10000, 0, 0.01)
```

```python
plt.plot(data_x, model(data_x, x.chain[-1]))
plt.plot(data_x, model(data_x, theta_true), 'k.')
```

```python
x.plot_histogram(0, 0.5, true=theta_true[0])
x.plot_histogram(1, 0.5, true=theta_true[1])
x.plot_histogram(2, 0.5, true=1.0)
```
